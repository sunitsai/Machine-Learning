# numpy with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt


t = np.arange(0.0,2.0,0.01)
s = np.cos(2*np.pi*t)

plt.plot(t,s)
plt.grid()
plt.xlabel("time(T)")
plt.ylabel("Voltage(mV)")
plt.title("Cosine wave plot(cos(x))")
plt.show()
```

# Subplot in Matplot

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,2.0)

y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(2,1,1)
plt.plot(x1,y1)

plt.title("Subplot-1")
plt.xlabel("x1")
plt.ylabel("Amp(y1)")
plt.grid()

# 2nd Subplot


plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.title("Subplot-2")
plt.xlabel("x2")
plt.ylabel("Amp(y2)")
plt.grid()

plt.show()


```

# MAtplotlib with CSV and Text File

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open("test.csv","r") as csvfile:
    plots = csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])

plt.plot(x,y,label="File")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Test Graph")
plt.legend()
plt.show()
```
